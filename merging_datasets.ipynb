```python
import pandas as pd

# Leggi il file CSV
df = pd.read_csv('backend/public/csv/tmqmsalen.csv')

# Dividi il dataset in base alla lunghezza del nome nella colonna 'Molecule'
df_short = df[df['Molecule'].str.len() <= 7]
df_long = df[df['Molecule'].str.len() > 7]

# Mostra le dimensioni dei due dataset
print(f"Dataset con nomi <= 7 caratteri: {len(df_short)} righe")
print(f"Dataset con nomi > 7 caratteri: {len(df_long)} righe")
```

```python
frames = [df_short, df2]
```
